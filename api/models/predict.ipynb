```python
first_pick = ["1528.jpg","1569.jpg"] # pilihan item pertama
# first_pick = ["1528.jpg"]
def predict(item):
    category = ["topWear","bottomWear"]
    row_index = []
    hm = {}
    for i in item:
        row_index.append(styles_df.index[styles_df['filename'] == i].tolist()[0])
    # print(row_index)
    for count,i in enumerate(row_index):
        dist, index = loaded_model.kneighbors(X=X[i,:].reshape(1,-1))
        hm2 = {}
        similar_item = []
        for j in range(10):
            similar_item.append({"id":styles_df.loc[index[0][j],"filename"],
                            "link":"None"})
        hm[f"{category[count]}"] = similar_item
    return hm
        
# hm = predict(first_pick)
# print(hm)
    
```

```python
input = {
    "gender":"Men",
    "usage":"Casual",
    "baseColour":"Blue"
}

def predict_first(input,df):

  df_baju = df[df["subCategory"]!="Bottomwear"]
  df_celana = df[df["subCategory"]=="Bottomwear"]

  df_baju = df_baju[df_baju["gender"]==input["gender"]]
  df_baju = df_baju[df_baju["usage"]==input["usage"]]
  df_baju = df_baju[df_baju["baseColour"]==input["baseColour"]]

  df_celana = df_celana[df_celana["gender"]==input["gender"]]
  df_celana = df_celana[df_celana["usage"]==input["usage"]]
  df_celana = df_celana[df_celana["baseColour"]==input["baseColour"]]
  # print((df["filename"]))
  # print(len(df))
  return [df_baju["filename"].sample(n=1).tolist()[0], df_celana["filename"].sample(n=1).tolist()[0]]


item1 = predict_first(input,styles_df)
predict(item1)
```
